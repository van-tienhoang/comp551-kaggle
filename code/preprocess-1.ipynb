```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import measure

```

```python
image1 = train_data.iloc[702, 1].reshape(100,100)
contours = measure.find_contours(image1, 0.8)

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(image1, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image1)
```

```python
max_lc = 1
cnt = None
for n, contour in enumerate(contours):
    l_c = len(contour)
    if l_c > max_lc:
        max_lc = l_c
        cnt = contour
        
plt.plot(cnt[:, 1], cnt[:, 0], linewidth=2)   
        
```
