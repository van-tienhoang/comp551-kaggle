```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import measure

```

```python
contours = measure.find_contours(image2, 0.8)

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(image2, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
```

```python
max_lc = 1
cnt = None
for n, contour in enumerate(contours):
    l_c = len(contour)
    if l_c > max_lc:
        max_lc = l_c
        cnt = contour
        
plt.plot(cnt[:, 1], cnt[:, 0], linewidth=2)   
        
```
