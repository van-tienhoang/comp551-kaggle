```python
import numpy as np
from copy import deepcopy
from PIL import Image

from pathlib import Path

mypath = Path().absolute()
print(mypath)
```

```python
train = np.load('../../Kaggle_data/train_images.npy', encoding = 'bytes')
np.set_printoptions(threshold=np.inf)
#np.set_printoptions(threshold=5)
```

```python
train_ex = train[2083][1] #INPUT ONE TRAINING EXAMPLE HERE
image_ex = deepcopy(train_ex)
train_ex = train_ex.ravel()
image_ex = image_ex.ravel()
image_ex.resize((100,100))
im = Image.fromarray(image_ex)
im.show()
train_ex = np.reshape(train_ex, (-1, 100))
```

```python
#TO DO:
#check if threshold even has any candidates. if not, loop until you get atleast one and then subtract by 10 (or whatever)

length = 30 #vertical, works well
# length = 30 #vertical
width = length #horizontal, works well
# width = 35
total_length = 100
total_width = 100
candidates_num = 0
candidates_list = []
density = np.zeros([total_length - (length - 1), total_width - (width - 1)])
#train_ex = [x+1 for x in train_ex]

for r in range(total_length - (length - 1)):
    for c in range(total_width - (width - 1)):
        subarray = train_ex[r:length + r , c:width + c]
        density[r][c] = np.mean(subarray)
        
max_density = np.max(density)
threshold = 0
candidates_min = 600

for i in range(1,50):
    threshold = max_density - i
    for r in range(total_length - (length - 1)):
        for c in range(total_width - (width - 1)):
            subarray = train_ex[r:length + r , c:width + c]
            density[r][c] = np.mean(subarray)
        
            if np.mean(subarray) > threshold:
                candidates_num = candidates_num + 1
                candidates_list.append([r,c])
         
    if candidates_num > candidates_min:
        break;
    
```

```python
#TODO: 
#change 255 to 0

strokes = 0
counts = 0
weights = np.zeros(np.shape(candidates_list)[0])
new_threshold = 210 #works well
# new_threshold = 255/2

def stroke_counter(row, column, array, length, width):
    matcher = 0
    matching = 0
    row_final = 0
    
    for new_row in range(row, length):
        if matcher == 1 and row_final == 0:
            row_final = 1
        elif matcher == 1 and row_final == 1:
            break;
        for new_column in range(column, width):
            if array[new_row][new_column] < new_threshold and matcher == 0:
                continue;
            elif array[new_row][new_column] >= new_threshold and matcher == 0:
                matcher = 1
#                 array[new_row][new_column] = 0 #added this recently
                continue;
            elif array[new_row][new_column] >= new_threshold and matcher == 1:
                matching = matching + 1
#                 array[new_row][new_column] = 0 #added this recently
                row_final = 0
                continue;
            elif array[new_row][new_column] < new_threshold and matcher == 1:
                column = new_column
                break;
    
    return matching

# fake_train_ex = deepcopy(train_ex) #added this recently

for idx, candidate in enumerate(candidates_list):
    r = candidate[0]
    c = candidate[1]
#     train_ex = deepcopy(fake_train_ex)
    candidate_array = train_ex[r:length + r, c:width + c] #added this recently
    
    for row in range(length):
        for column in range(width):
            if column <= (length - 2) and row <= (width - 2):
                if candidate_array[row][column] >= new_threshold and candidate_array[row][column+1] >= new_threshold:
                    counts = counts + stroke_counter(row = row, column = column, array = candidate_array, length = length, width = width)
                    strokes = strokes + 1
    
    weights[idx] = counts/strokes
    counts = 0
    strokes = 0

chosen_candidate = candidates_list[np.argmax(weights)]
array = train_ex[chosen_candidate[0]:length + chosen_candidate[0] , chosen_candidate[1]:width + chosen_candidate[1]]
array = array.ravel()
array.resize((length,width))
im = Image.fromarray(array)
im.show()

            
    
```
